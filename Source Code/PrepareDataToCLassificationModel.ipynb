```python
# Read video
import os

def get_video_paths(folder_path):
    video_paths = []
    valid_extensions = ['.mp4', '.avi', '.mkv']

    for root, dirs, files in os.walk(folder_path):
        for file in files:
            _, extension = os.path.splitext(file)
            if extension.lower() in valid_extensions:
                video_path = os.path.join(root, file)
                video_paths.append(video_path)

    return video_paths


folder_path = '/content/drive/MyDrive/DemoPoseAndClassifier/VideoFrame'

video_paths = get_video_paths(folder_path)
for path in video_paths:
    print(path)

print('\n # video:',len(video_paths))
```

```python
# Calculate the number of frame in video
import cv2

def get_frame_count(video_path):
    cap = cv2.VideoCapture(video_path)
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    cap.release()
    return frame_count
```

```python
# Get file name
def Get_DataFileName(provenance):
   filename_ext = provenance.split('/')[-1]
   filename = filename_ext.split('.')[0]
   return filename
```

```python
# Prepare data to Classifier Model
import json
import csv

def json_file_to_object(json_file):
    with open(json_file, 'r') as file:
        json_data = file.read()
        json_object = json.loads(json_data)

    return json_object

file_path = '/content/Tray164.csv'

Allpoints=[]

for path in Json_paths:

  json_object = json_file_to_object(path)
  mynode=json_object['nodes']
  provenance=json_object['provenance']
  mydataname=Get_DataFileName(provenance['data_path'])

  for label in json_object['labels']:
    myframename=label['frame_idx']
    for instances in label['_instances']:
      mypoint=['']*len(decnodes)
      for index, point in instances['_points'].items():

        mypoint[decnodes[mynode[int(index)]['name']+'_X']]=point['x']
        mypoint[decnodes[mynode[int(index)]['name']+'_Y']]=point['y']
        #print(index,"==>",decnodes[mynode[int(index)]['name']+'_X'],"////",decnodes[mynode[int(index)]['name']+'_Y'])

        #mypoint[decnodes[nodes[int(index)]+'_X']]=point['x']
        #mypoint[decnodes[nodes[int(index)]+'_Y']]=point['y']

      mypoint[-2]=mydataname
      mypoint[-1]=myframename
      Allpoints.append(mypoint)

with open(file_path, 'w', newline='') as csvfile:
  writer = csv.writer(csvfile)
  writer.writerow(decnodes)
  writer.writerows(Allpoints)
csvfile.close()

print(decnodes)
print(Allpoints)
```
